6. Explain the concept of generators in Python and how they are defined?
   - Generators are a special type of iterable in Python, used to create iterators in a concise and memory-efficient way. Instead of generating all the values at once (like lists), generators yield one value at a time, only as needed.
   - Defining Generators
  - Generators can be defined in two ways:
  *1. Using yield in a Function
  - A generator function is defined like a regular function but uses the yield keyword to produce values one at a time.
  - When yield is called, the function's state is saved, and it can be resumed later.
  -# Example
  - def count_up_to(n):
   - count = 1
   - while count <= n:
   -    yield count  # Produces the next value
   -   count += 1
 *2. Generator Expressions
  - A compact way to create generators, similar to list comprehensions but with parentheses instead of square brackets.
  -# Example
  - squares = (x**2 for x in range(5))
  - Key Characteristics of Generators
  *1. Lazy Evaluation:
  - Generators produce items one at a time, making them memory-efficient, especially for large datasets.
  *2. Statefulness:
  - A generator function "remembers" its execution state between calls, - - which allows it to resume exactly where it left off.
  *3. Exhaustion:
  - Once a generator has produced all its values, it cannot be restarted. - A new generator instance must be created.
  - Using Generators
  - Generators can be used with loops, explicitly fetched using next(), or converted to other iterables like lists.

'Filter' constructs a list from elements of an iterable for which a function returns true. 'Reduce' applies a rolling computation to sequential pairs of values in a list and returns a single result.Map, filter, and reduce are three powerful functions in Python that allow you to manipulate and transform data in efficient and concise ways. These functions are part of the functional programming paradigm, which emphasizes immutability and the use of higher-order functions to create more modular and reusable code.
  - Map, filter, and reduce are three powerful functions in Python that allow you to manipulate and transform data in efficient and concise ways. These functions are part of the functional programming paradigm, which emphasizes immutability and the use of higher-order functions to create more modular and reusable code.   
  *1. map():
  - Purpose: Applies a given function to each element of an iterable (e.g., a list, tuple, or string) and returns a new iterable containing the results.
  - Syntax: map(function, iterable).
  -# Example
  - numbers = [1, 2, 3, 4]
   - squared = map(lambda x: x**2, numbers)
   - print(list(squared))
   - output [1, 4, 9, 16]

   - Purpose: Filters elements from an iterable based on a given function that returns True or False. It creates a new iterable containing only the elements that satisfy the condition.
   -  Syntax: filter(function, iterable).

   - print(list(evens))
   - output [2, 4, 6]

  - Purpose: Applies a function to a sequence of elements, reducing them to a single value. It takes two elements at a time, applies the function to them, and then uses the result with the next element, and so on.

  - List comprehensions and generator expressions can often be used as alternatives to map() and filter(), offering a more concise syntax in some cases.

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False  # 0 and 1 are not prime numbers
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False  # n is divisible by i, so it's not prime
    return True  # n is prime

# Check numbers from 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms  # Set the maximum number of terms
        self.count = 0              # Initialize the count of terms generated
        self.a, self.b = 0, 1       # Initialize the first two Fibonacci numbers

    def __iter__(self):
        return self  # Return the iterator object itself

    def __next__(self):
        if self.count < self.max_terms:  # Check if we have more terms to generate
            result = self.a               # Store the current Fibonacci number
            self.a, self.b = self.b, self.a + self.b  # Update to the next Fibonacci numbers
            self.count += 1               # Increment the count
            return result                 # Return the current Fibonacci number
        else:
            raise StopIteration           # Stop iteration when max_terms is reached

# Example usage:
fib = FibonacciIterator(10)  # Create an instance for 10 terms
for num in fib:              # Iterate through the Fibonacci numbers
    print(num)               # Print each Fibonacci number
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2_yield(n):
    for i in range(n + 1):
        yield 2 ** i

for power in powers_of_2_yield(5):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
file_content = [
    "Hello, this is the first line.",
    "This is the second line.",
    "And this is the third line.",
    "This is the fourth line.",
]


def read_file_line_by_line(content):
    for line in content:
        yield line.strip()


for line in read_file_line_by_line(file_content):
    print(line)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    """Convert Celsius to Fahrenheit."""
    return (celsius * 9/5) + 32  # Formula to convert Celsius to Fahrenheit

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100, -40]

# Use map() to apply the conversion function to each temperature in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the results
print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.
def is_not_vowel(char):
    """Check if the character is not a vowel."""
    return char.lower() not in 'aeiou'  # Return True if the character is not a vowel

# Given string
input_string = "Hello, World!"

# Use filter() to remove vowels from the string
filtered_string = ''.join(filter(is_not_vowel, input_string))  # Join the filtered characters back into a string

# Print the result
print("Original string:", input_string)
print("String without vowels:", filtered_string)
```

```python
"""11.Imagine an accounting routine used in a book shop. It works on a list with sublists, which look like this:
Write a Python program, which returns a list with 2-tuples. Each tuple consists of the order number and the
product of the price per item and the quantity. The product should be increased by 10,- € if the value of the
order is smaller than 100,00 €.
Write a Python program using lambda and map."""

orders = [
    [101, 4,40.00],
    [202, 5, 56.80],
    [303, 3, 32.95],
    [404, 3, 24.99],
]
def calculate_order(order):
    order_number, price_per_item, quantity = order
    total_price = price_per_item * quantity
    if total_price < 100:
        total_price += 10
    return (order_number, total_price)
result = list(map(lambda order: calculate_order(order), orders))

print(result)
```
